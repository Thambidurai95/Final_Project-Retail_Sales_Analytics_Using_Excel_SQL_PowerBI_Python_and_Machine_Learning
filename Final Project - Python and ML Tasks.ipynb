```python
# Task 3 - RFM Calculation for Customers
    # Compute Recency, Frequency, and Monetary values for each customer.
    # Recency: Days since last order
    # Frequency: Number of orders
    # Monetary: Total value of all purchases
```

```python
# Merging orders and order_items datasets
pd.merge(orders_df,order_items_df,on='order_id')
```

```python
# Storing it in a variable
merged_orders = pd.merge(orders_df,order_items_df,on='order_id')
```

```python
# Merging merged_orders and customers datasets
pd.merge(merged_orders,customers_df,on='customer_id')
```

```python
# Storing it in a variable
merged_dataset = pd.merge(merged_orders,customers_df,on='customer_id')
```

```python
# Finding the most recent order data and adding one extra date to the max date to freeze the date for date difference
merged_dataset['order_date'].max() + pd.Timedelta(days=1)
```

```python
# Storing it in a variable
freezed_date = merged_dataset['order_date'].max() + pd.Timedelta(days=1)
```

```python
# Grouping the customer_id in the merged_dataset and calculating RFM through aggregation
merged_dataset.groupby('customer_id').agg({
'order_date':lambda x: (freezed_date - x.max()).days,     # recency
'order_id':'nunique',                                     # frequency
'total_price':'sum'                                       # monetary
}).reset_index()
```

```python
# Storing it in a variable
rfm = merged_dataset.groupby('customer_id').agg({
'order_date':lambda x: (freezed_date - x.max()).days,     # recency
'order_id':'nunique',                                     # frequency
'total_price':'sum'                                       # monetary
}).reset_index()
```

```python
# Renaming the columns for rfm dataset
rfm.columns = ['customer_id','recency','frequency','monetary']
```

```python
# Retrieving rfm dataset with newly updated columns
rfm
```

```python
# Retrieving top 5 datas from rfm
print(rfm.head())
```

```python
# Customer Segmentation using KMeans (Basic Clustering)
    # Use RFM data from above.
    # Normalize using MixMaxScaler.
    # Apply KMeans to segment customers into 3â€“4 groups.
```

```python
# Importing the MinMaxScaler library
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scaling method to normalize the range between 0 and 1
MinMaxScaler()
```

```python
# Calculates min-max values of each feature(.fit) and scales the values in to the range [0,1] based on the min-max normalization formula (.transform)
MinMaxScaler().fit_transform(rfm[['recency','frequency','monetary']])
```

```python
# Storing it in a variable
rfm_normalized = MinMaxScaler().fit_transform(rfm[['recency','frequency','monetary']])
```

```python
# Importing KMeans library
from sklearn.cluster import KMeans
```

```python
# Using KMeans library, creating customer segmentaion in to 4 groups by freezing the output with random state of 40 
KMeans(n_clusters=4,random_state=40)
```

```python
# Storing it in a variable
kmeans = KMeans(n_clusters=4,random_state=40)
```

```python
# Train (.fit) and label together (.predict) to cluster
kmeans.fit_predict(rfm_normalized)
```

```python
# Storing it in a variable
rfm['segment'] = kmeans.fit_predict(rfm_normalized)
```

```python
# Retrieving the segment and clusters from rfm for each customers
rfm
```

```python
# Retrieving rfm[segment] data value counts for each segment
print(rfm['segment'].value_counts())
```

```python
# Scatter plotting for rfm
sns.scatterplot(x='recency',y='monetary',hue='segment',data=rfm)
plt.title('Customer Segments by Recency and Monetary Value')
plt.show()
```
